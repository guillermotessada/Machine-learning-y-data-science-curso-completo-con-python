## Arrays

Un **array** es una estructura de datos que consiste en una colección de elementos (valores o variables), cada uno identificado por al menos un índice o clave. Un array se almacena de modo que la posición de cada elemento se pueda calcular a partir de su tupla de índice mediante una fórmula matemática. El tipo más simple de array es un array lineal, también llamado array unidimensional.Un array es modificable, como sucede para las listas.

En numpy:
* Cada dimensión se denomina **axis**
* El número de dimensiones se denomina **rank**
* La lista de dimensiones con su correspondiente longitud se denomina **shape**
* El número total de elementos (multiplicación de la longitud de las dimensiones) se denomina **size**
* 


```python
# Array cuyos valores son todos 0 ''
a = np.zeros((2,4)) # 2 filas y 4 columnas
print('array de zeros')
print (a)
print('forma array = ', a.shape)
```

_**a**_ es un array:
* Con dos **axis**, el primero lineas de longitud 2 y el segundo columnas de longitud 4
* Con un **rank** igual a 2 -- dos dimensiones 
* Con un **shape** igual (2, 4) -- 2 filas y 4 columnas
* Con un **size** igual a 8 -- contiene 8 elementos
* Son como dos listas anidadas

```python
# rank
print('dimensiones array = ', a.ndim)
a
```

```python
# numero elementos del array
print('numero elementos array = ', a.size)
a
```

```python
# Array cuyos valores son todos 1
np.ones((2,3,4))
```

```python
# Inicializacion del array utilizando un lista de Python
b = np.array([1,2,3]) # la lista esta entre corchetes
# la lista tiene 3 valores y 1 dimension
print (b) # array
print('forma array = ', b.shape) # 1 linea y 3 columnas
print('dimensiones array = ', b.ndim) # 1 dimensione
```

```python
# Inicializacion del array utilizando dos o mas listas listas de Python
b = np.array([[1,2,3],[4,5,6]]) # la lista esta entre corchetes
print (b) # array
print('forma array = ', b.shape) # 2 lineas y 3 columnas
print('dimensiones array = ', b.ndim) # 2 dimensiones 
```

```python
# Creación del array utilizando una función basada en rangos
# (minimo, maximo, número elementos del array)
c = np.linspace(0,6,10) # crea un array con inicio en 0, con maximo 6 y con 10 elementos
c
print (c) # array
print('forma array = ', c.shape) # 1 linea y 6 columnas
print('dimensiones array = ', c.ndim) # 1 dimensiones 
print('numero elementos array'c.size) # numero elementos array
```

```python
# Inicialización del array con valores aleatorios
d=np.random.rand(2,3,4) 
print (d) # array
print('dimensiones array = ', d.ndim) # 3 dimensiones 
print('numero elementos array = ', d.size) # numero elementos array
```

```python
# Inicialización del array con valores aleatorios conforme a una distribución normal
e = np.random.randn(2,3,4)
print (e) # array
print('dimensiones array = ', e.ndim) # 3 dimensiones 
print('numero elementos array = ',e.size) # numero elementos array
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
g = np.random.randn(1000000)
plt.hist(g, bins = 200)
print (g) # array
print('forma array = ', g.shape) 
print('dimensiones array = ', g.ndim) # 1 dimensiones 
print('numero elementos array = ', g.size) # numero elementos array
plt.show
```

## Acceso a los elementos de un array

### Array unidimensional

```python
# Creación de un Array unidimensional
array_uni = np.array([1,3,5,7,9,11])
print('forma array = ', array_uni.shape)
print(array_uni)
print('tipo objeto = ', type(array_uni))
```

```python
# Accediendo al quinto elemento del Array como lo que se usa en las listas
array_uni[4]
```

```python
# Accediendo al tercer y cuarto elemento del Array slicing
array_uni[2:4]
```

```python
# Accediendo a los elementos 0, 3 y 5 del Array
# el ultimo valor es el step va de 2 en dos
array_uni[0::2]
```

### Array multidimensional

```python
# Creación de un Array multidimensional
array_multi = np.array([[1,2,3,4],[5,6,7,8]])
print('forma array = ',array_multi.shape)
print(array_multi)
```

Para acceder a los elementos de un array la sintaxis:
* para un array unidimensional es **array[index_posicion]**
* para una array multidimensional es **array[filan1:filan2,index_posicion]**
* para una array multidimensional es **array[filan_n,index_posicion]**

```python
# Accediendo al cuarto elemento del Array esto es diferente con las listas
array_multi[0,3]
```

```python
# Accediendo a la segunda fila del Array
print('acceso primera fila')
print(array_multi[0])
print('acceso segunda fila')
print(array_multi[1])
```

```python
# Accediendo al tercer elemento de las dos primeras filas del Array
print('primer modo')
print(array_multi[:,2])
print('segundo modo')
print(array_multi[0:2,2])
```

## Modificación de un Array

```python
# Creación de un Array unidimensional inicializado con el rango de elementos 0-27
# uso del metodo arange(primernumero, ultimonumero) el ultimo no es incluido
# el reshape tiene que t
array1 = np.arange(0,28)
array1
```

```python
# Cambiar las dimensiones del Array y sus longitudes
# el reshape tiene que cumplir con el numero de elementos
# en este caso son 28 = 7*4
array1.shape = (7,4)
print(array1)
print('forma array = ', array1.shape)
```

Nota:
* Importante: Modificaciones en un Array, modificaran el array original
es conveniente crear un nuevo array desde el original


```python
# Modificación del nuevo Array devuelto
array2 = array1.reshape(4,7)
array2
```

```python
array1
```

```python
# modifica de un elemento en un array
# Importante: El nuevo array apunta a los mismos array el de origen y el modificado
# el tercer elemento de la fila 0 es ahora 20
array2[0,3] = 20
array2
```

```python
# la modifica en el array2 tiene efecto sobre el array1
array1
```

```python
# Desenvuelve el array es decir devuelve un nuevo array de una sola dimension
array3 = array1.ravel()
print(array3)
print('forma array = ' ,array3.shape)
```

```python
# para crear una copia de un array se usa np.copy
array4 = np.copy(array3)
array4
```

```python
# modifico la copia y no tiene efecto sobre la original
array4[3] = 35
print(array4)
print(array3)
```

## Operaciones aritméticas con Arrays

```python
# Creación de dos Arrays unidimensionales
#np.arange(inicio, fin, step)
# los dos array tienen la misma dimension y el mismo numero de valores
array1 = np.arange(2,18,2)
print(array1)
print('dimension array1 = ', np.ndim(array1))
print('numero elementos array1 = ', np.size(array1))
array2 = np.arange(8)
print(array2)
print('dimension array2 = ', np.ndim(array2))
print('numero elementos array2 = ', np.size(array2))
```

```python
# Suma
# suma es posible solo si los array tienen las mismas dimensiones o forma (shape)
# 1era suma 0+2 ; 2da suma 1+4; ..
sumaarray = array1 + array2
sumaarray
```

```python
# Resta
# resta es posible solo si los array tienen las mismas dimensiones o forma (shape)
restaarray12 = array1-array2
print(restaarray12)
restaarray21 = array2-array1
print(restaarray21)
```

```python
# Multiplicacion
# Importante: No es una multiplicación de matrices
# 1era mult 2*0; 2da mult 4*1;
multarray = array1 * array2
multarray
```

## Broadcasting

Si se aplican operaciones aritméticas sobre Arrays que no tienen la misma forma (shape) Numpy aplica un propiedad que se denomina Broadcasting.

```python
# Creación de dos Arrays unidimensionales
array1 = np.arange(5)
print(array1)
print('forma array1 = ', array1.shape)
array2 = np.array([3])
print(array2)
print('forma array2 = ', array2.shape) 
```

```python
# Suma de ambos Arrays con forma diferente
# suma el elemento de array menor al mayor
sumaarrayd = array1 +array2
sumaarrayd
```

```python
# Creación de dos Arrays multidimensional y unidimensional 
array1 = np.arange(6)
array1.shape = (2,3)
print( array1)
print('forma array1 = ', array1.shape)
array2 = np.arange(6,18,4)
print(array2)
print('forma array2 = ', array2.shape)
```

```python
# Suma de ambos Arrays
sumaarrayfd = array1 + array2
sumaarrayfd
```
